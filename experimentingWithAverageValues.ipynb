```python
targetDF = pd.DataFrame()
targetDF['Time'] = pd.date_range(start=startTime, end=endTime, freq='1min')

prepdList = []
for datF in prepList:
    datF['Date'] = datF['Time'].dt.date
    datF['Hour'] = datF['Time'].dt.hour
    datF['Minute'] = datF['Time'].dt.minute
    datF.drop(['Time'], axis = 1, inplace=True)

    datNF = datF.groupby(['Date', 'Hour', 'Minute'])[datF.columns[0]].mean()
    datNF = datNF.reset_index()
    datNF['Time'] = datNF['Date'].astype(str) + ' ' + datNF['Hour'].astype(str) + ':' + datNF['Minute'].astype(str) + ':00'
    datNF['Time'] = pd.to_datetime(datNF['Time'], format="%Y-%m-%d %H:%M:%S")
    datNF.drop(['Date', 'Hour', 'Minute'], axis = 1, inplace=True)

    targetDF = pd.merge(targetDF, datNF, on='Time', how='left')
```

```python
import numpy as np
from sklearn.svm import SVR
from sklearn.preprocessing import PolynomialFeatures

def customInterpolation(tdf, col):
    df = tdf.copy()
    retdf = tdf.copy()
    
    for index, row in df.iterrows():
        if pd.isnull(row[col]):
            nearest = []
            cur_ind = index
            
            while len(nearest) < 7 and cur_ind >= 0:
                if pd.notnull(df.iloc[cur_ind][col]):
                    nearest.append([cur_ind ,df.iloc[cur_ind][col]])
                cur_ind -= 1
            
            cur_ind = index + 1

            val = len(nearest)

            while len(nearest) < val + 7 and cur_ind < df.shape[0]:
                if pd.notnull(df.iloc[cur_ind][col]):
                    nearest.append([cur_ind, df.iloc[cur_ind][col]])
                cur_ind += 1

            nearest = sorted(nearest, key=lambda x: abs(x[0] - index))[:7]
            
            X_known = np.array([el[0] for el in nearest]).reshape(-1, 1)
            y_known = np.array([el[1] for el in nearest])

            poly = PolynomialFeatures(degree=2)
            X_known_poly = poly.fit_transform(X_known)

            model = SVR(kernel='poly', degree=2)
            model.fit(X_known_poly, y_known)

            X_unknown = np.array(index).reshape(-1, 1)    
            X_unknown_poly = poly.transform(X_unknown)  

            predicted = model.predict(X_unknown_poly)
            retdf.at[index, col] = predicted[0]              

    return retdf
```

```python
for column in targetDF.columns:
    if column != 'Time':
        targetDF = customInterpolation(targetDF, column)
targetDF
```
